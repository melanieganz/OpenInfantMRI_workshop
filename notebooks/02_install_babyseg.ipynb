# Installing babyseg CPU locally

Read documentation here: https://github.com/mu40/babyseg/

## Step 1: Download the wrapper script to the babyseg directory.

https://github.com/mu40/babyseg/blob/main/docker/README.md
The wrapper scripts faciliates setup through containers.
Ensure you have a container downloaded, e.g. Docker, Podman, Apptainer, or Singularity.

```python
import os
babyseg_path = "../babyseg" 
# download the wrapper into the babyseg_path

os.makedirs(babyseg_path,exist_ok=True)
```

## Step 2: Run the wrapper script. 
This will build the babyseg installation into your babyseg directory

The wrapper can be run in the terminal by simply typing: ``python ../babyseg/wrapper.py``
Alternatively you can run the code below. This runs bash through subprocess

```python
import subprocess
import sys
from pathlib import Path

wrapper = (Path("..") / "babyseg" / "wrapper.py").resolve()
if not wrapper.exists():
    raise FileNotFoundError(f"wrapper.py not found at: {wrapper}")

result = subprocess.run(
    [sys.executable, str(wrapper)],
    text=True,
    capture_output=True,
    check=False,
 )

print(result.stdout)
if result.returncode != 0:
    print(result.stderr)
    raise RuntimeError(f"wrapper.py failed with exit code {result.returncode}")
```

## Step 3: Finalise installation and make the babyseg command executable 

This should print out the help for babyseg installation. Read the installation

```python
import os
import subprocess
import sys
from pathlib import Path

babyseg_home = (Path("..") / "babyseg").resolve()
wrapper = babyseg_home / "wrapper.py"
if not wrapper.exists():
    raise FileNotFoundError(f"wrapper.py not found at: {wrapper}")

env = os.environ.copy()
env["BABYSEG_HOME"] = str(babyseg_home)
env["BABYSEG_SIF"] = str(babyseg_home)

# Cross-platform: call the wrapper directly with the active Python interpreter.
result = subprocess.run(
    [sys.executable, str(wrapper), "-h"],
    env=env,
    text=True,
    capture_output=True,
    check=False,
 )

if result.stdout:
    print(result.stdout)
if result.returncode != 0:
    if result.stderr:
        print(result.stderr)
    raise RuntimeError(f"babyseg help command failed with exit code {result.returncode}")
```
